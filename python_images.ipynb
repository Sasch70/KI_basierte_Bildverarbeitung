### Matplotlib
Bilder einladen und anzeigen.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

plt.subplot(121) # nrows,ncols,index
img = mpimg.imread('data/greenscreen.jpg') #Format (1944,1555,3)
plt.imshow(img)

plt.subplot(122)
back = mpimg.imread('data/background.jpg')
plt.imshow(back)

print(img.shape)
```

```python
# Bilder spiegeln
plt.imshow(np.flip(img,axis=0))

# oder mit flipud
plt.imshow(np.flipud(img))
```

```python
plt.subplot(121) 
plt.imshow(np.rot90(img))

plt.subplot(122) 
plt.imshow(np.rot90(img,k=3)) #k: Anzahl Drehungen
```
